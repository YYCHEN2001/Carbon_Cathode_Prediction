```python
import pandas as pd

data = pd.read_csv('dataset_processed.csv')

data.head()
```

```python
from sklearn.preprocessing import StandardScaler

data_1 = data.drop(columns=['Elyte', 'CC'])

scaler = StandardScaler()
data_scaled_array = scaler.fit_transform(data_1)

data_scaled = pd.DataFrame(data_scaled_array, columns=data_1.columns)
data_scaled
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 设置图形风格
sns.set(style="whitegrid")

# 绘制小提琴图
sns.violinplot(data=data_scaled, inner="box", linewidth=2)

# 设置字体和图像尺寸
plt.rcParams.update({
    'font.family': 'Times New Roman',
    'font.size': 32,
    'font.weight': 'bold',
    'figure.figsize': (16, 9)  # 设置图像尺寸为10x10英寸
})

# 设置标题和标签
plt.title('Feature Distribution Violin Plot',fontweight='bold', pad=20)
plt.xlabel('Features', fontweight='bold')
plt.ylabel('Standardized Value', fontweight='bold')

# 设置图形边界的宽度和可见性
for spine in plt.gca().spines.values():
    spine.set_visible(True)
    spine.set_linewidth(2.5)
    spine.set_color('black')
    
# 保存图像，背景透明，紧凑布局
plt.savefig('../output/figures/violin.png', bbox_inches='tight', transparent=True)
plt.show()
```
